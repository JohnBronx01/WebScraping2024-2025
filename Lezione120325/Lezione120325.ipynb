## Le espressioni regolari (REGEX) in Python

* un semplice linguaggio utilizzato per descrivere **patterns** che si ricercano all'interno di un contenuto
* utilizzate in diversi linguaggi di programmazione
* **JAVA**, **Python**, **JAVASCRIPT** e molti altri
    * https://blog.teamtreehouse.com/regular-expressions-10-languages
* in **Python**: libreria **re** 
* usate con **BeautifulSoup**, permettono di eseguire ricerche più complesse ed estrarre dati basandoti su **patterns**. 
* particolarmente utili quando gli elementi **HTML** hanno attributi dinamici o variabili, perchè permettono  un modo più flessibile per individuarli.


## Le espressioni regolari (REGEX) in Python

```python
import re
html = "<html><head><title>Scraping</title></head></html>"
title_search = re.search("<title>(.*?)</title>", html)
title = title_search.group(1)
print(title)
```

* funzione **re.search** ricerca un particolare pattern all'interno di una **stringa**
* pattern **\<title\>(.*?)\</title\>**
  * **\<title\>** è una stringa, il pattern ricercato dovrà contenere esattamente questa stringa iniziale
  * **(.*?)**: denota un gruppo di caratteri in cui
    * il primo . può essere qualsiasi
    * successivamente si ripetono 0 o più caratteri 
  * **\<\/title\>**:è una stringa, il pattern ricercato dovrà contenere esattamente questa stringa finale
    

## Le espressioni regolari (REGEX) in Python: re.compile()

* utilizzato per compilare una espressione regolare fornita come stringa in un oggetto **pattern regex (re.Pattern)**. 
* successivamente, è possibile usare questo oggetto pattern per cercare matching  all'interno di stringhe 
diverse con **re.match()** o **re.search()**.
* maggiore efficienza, perchè la espressione regolare viene compilata in una struttura interna una sola volta e poi può essere riutilizzata più volte

## Le espressioni regolari (REGEX) in Python: re.compile()

* numeri di cellulare:**10 cifre**, inizino con **+91**
* pattern = **\+91[0-9]{10}**
    * simbolo **+** va trattato come un carattere e non come un simbolo speciale
    * il pattern deve iniziare esattamente con il valore  **91**
    * ci devono essere poi cifre tra **0** e **9** e ci devono essere **10** cifre

```python
import re
pattern = r"\+91[0-9]{10}"
student1 = "Matteo: +917363536475"
student2 = "Anna: +917468890483, +399846274959"
student3 = "Luca: +397483659048, +918754627894"
#creare un oggetto regex per il pattern
find_mobile = re.compile(pattern) 
#verifica i numeri telefonici
print("Matteo Mobile Numbers:", find_mobile.findall(student1))
print("Anna Mobile Numbers:", find_mobile.findall(student2))
print("Luca Mobile Numbers:", find_mobile.findall(student3))
```

## Le espressioni regolari per lo scraping: estrarre pattern

```python
import urllib.request
import re
response = urllib.request.urlopen("https://pypi.org/")
html = response.read().decode("utf-8")
matches = re.search("([0-9,]+) projects", html)
print (type(matches))
projects = matches.group(1)
```

* projects: la stringa ricercata deve terminare con la parola **projects**, preceduta da uno spazio
* **matches.group(1)** viene stampato solo il primo pattern restituito
* nessun controllo sul numero di virgole  contenute nel pattern

## Le espressioni regolari per lo scraping: estrarre testo


```python
from bs4 import BeautifulSoup
import re
html = '''
<div class="product">
    <h2>Product Title</h2>
    <span class="price">$19.99</span>
    <span class="sku">SKU: 12345</span>
</div>
<div class="product">
    <h2>Another Product</h2>
    <span class="price">$24.99</span>
    <span class="sku">SKU: 67890</span>
</div>
'''
```

## Le espressioni regolari per lo scraping: estrarre testo

* il prezzo:
  * deve iniziare con un **$**
  * deve contenere una sequenza di almeno una cifra da **0** a **9** e un punto
<code>
r'\$[\d\.]+'
</code>
* **group()**
    * restituisce il pattern per cui ha trovato la corrispondenza nella stringa data

```python
soup = BeautifulSoup(html, 'html.parser')
# Use a regular expression to find all prices
prices = soup.find_all('span', class_='price')
for price in prices:
    price_value = re.search(r'\$[\d\.]+', price.text)
    print(type(price_value))
    if price_value:
        print(price_value.group())
```

## Importare le librerie e preparare la url

* la **url** è composta da due parti diverse
    * una parte fissa
    * una parte variabile

```python
import requests
from bs4 import BeautifulSoup
from csv import writer
from time import sleep
from random import choice
 # la lista in cui memorizzeremo tutte le quote
all_quotes = []
 # questa parte della URL rimane fissa
base_url = "http://quotes.toscrape.com/"
 # questa parte della url indica la pagina visitata, all'inizio la prima pagina
# il numero di pagina verrà incrementato mano a mano che il codice scorrerà le pagine
url = "/page/1"
```

## Collezionare le quotes in una lista

```python
while url:   
    # concateniamo la url base con la url di pagina e effettuiamo la richiesta
    # costruzione di una stringa con i valori di due variabili   
    res = requests.get(f"{base_url}{url}")
    print(f"Now Scraping {base_url}{url}")   
    soup = BeautifulSoup(res.text, "html.parser") 
    # estraiamo le "quotes" da questa pagina
    quotes = soup.find_all('div', class_="quote")
    for quote in quotes:
        all_quotes.append({
            "text": quote.find(class_="text").get_text(),
            "author": quote.find(class_="author").get_text(),
            "bio-link": quote.find("a")["href"]
        })
    next_btn = soup.find(class_="next")
    url = next_btn.find("a")["href"] if next_btn else None
print(len(all_quotes))
print(all_quotes[1])
```

## Collezionare le quotes in una lista: osservazioni sul codice

     url = next_btn.find("a")["href"]
     </code>
* segue il link se diverso da **None**, altrimenti esce dal while, tutte le pagine sono state visitate

## Fase di gioco

```python
quote = choice(all_quotes)
remaining_guesses = 4
print("Here's a quote:  ")
print(quote["text"])
        
```

## Fase di Gioco

```python
guess = ''
while guess.lower() != quote["author"].lower() and remaining_guesses > 0:
    guess = input(
        f"Who said this quote? (Guesses remaining {remaining_guesses}) ")    
    if guess == quote["author"]:
        print("CONGRATULATIONS!!! YOU GOT IT RIGHT")
        break
    remaining_guesses -= 1    
    if remaining_guesses == 3:
        res = requests.get(f"{base_url}{quote['bio-link']}")
        soup = BeautifulSoup(res.text, "html.parser")
        birth_date = soup.find(class_="author-born-date").get_text()
        birth_place = soup.find(class_="author-born-location").get_text()
        print(
            f"Here's a hint: The author was born on {birth_date}, {birth_place}")
     
    elif remaining_guesses == 2:
        print(
            f"Here's a hint: The author's first name starts with: {quote['author'][0]}")
     
    elif remaining_guesses == 1:
        last_initial = quote["author"].split(" ")[1][0]
        print(
            f"Here's a hint: The author's last name starts with: {last_initial}")
    else:
        print(
            f"Sorry, you ran out of guesses. The answer was {quote['author']}")
```

## CSS: Introduzione

* il progetto iniziale di **HTML** prevedeva che gli sviluppatori si concentrassero prevalentemente sul contenuto delle pagine, lasciando decidere il layout ai browsers che ricevono tale contenuto

* nonostante questa idea iniziale, la realtà è che sono stati creati numerosi tag e proprietà riguardanti lo stile da applicare al contenuto da visualizzare
    * **font**, **center**, **align**, **color** **bold** **italic**
    * la situazione è risultata "fuori controllo" in pochi anni....

* la risposta del consorzio *W3C (World-Wide Web Consortium @ www.w3c.org)*
    * distinguere il **contenuto** dalla **presentazione** del contenuto
    * introduzione di un  **Style Sheet Language**, chiamato **CSS (Cascading Style Sheet)** per la definizione del layout dei contenuti
    * degli attributi di stile da *HTML*
    * linguaggi diversi per graphic designers e content providers

## Cosa è uno cascating style sheet?

* una collezione di **regole di stile**  applicabili a un **insieme selezionato** di **elementi HTML**
    * controlla il modo in cui il corrispondente elemento o i corrispondenti insiemi di elementi vengono visualizzati
    * il termine **cascading** significa che più regole di stile **a cascata** possono essere applicate allo stesso elemento
        **HTML**
    * il browser segue un certo **cascading order** per formattare quell'elemento
* è una lista di coppie chiave, valore, separate da punto e virgola

```python
small_example_css = """
      color: 'red';
      background-color: #ccc;
      font-size: 14pt;
      border: 2px solid yellow;
        """
print(small_example_css)
```

## Dove si trovano le dichiarazioni di stile?

* esistono tre possibilità
    * **inline** usando l'attributo **style** all'interno di un elemento **HTML**
        * **\<h1 style="color:blue;"\>A Blue Heading<\/h1>**
        * **\<p style="color:red;"\>A red paragraph.\</p\>**
    * **internal** usando un elemento **\<style\>** nella sezione **\<head\>** di un documento
    * in un file separato, che viene riferito attraverso un tag **\<link\>** inserito nella sezione **\<head\>** di un documento
        * questa è la migliore modalità di uso dei file **CSS**
        * quando il browser carica una pagina web, il browser effettua un'ulteriore richiesta **HTTP** per scaricare il file **CSS** ed applica le direttive di stile al documento
* nel primo caso è evidente quale è l'elemento a cui deve essere applicato lo stile
* negli altri due casi, la definizione di stile deve integrare anche l'informazione che indica l'elemento o gli elementi 
  **HTML** a cui la dichiarazione viene applicata
    * per questo vengono introdotti i **CSS selectors**

```python
css_selector_1 = """
/* Rule 1: Apply to the <body> element and possibly its descendants */
body {
  font-family: "Open Sans", Helvetica, Arial, sans-serif;
  font-size: 16px;            /* 16px = 12pt */
  margin: 5px 15px 5px 15px;  /* top right bottom left - no commas in between */
  padding: 0;
  background-color: #eee;    /* light gray, same as #eeeeee */
}
 
/* Rule 2: Apply to <h1> to <h6> elements */
h1, h2, h3, h4, h5, h6 {
  font-family: Quicksand, "Open Sans", Helvetica, Arial, sans-serif;
  color: red;
  font-weight: bold;
  text-align: center;
}
  
/* Rule 3: Specifically for <h2>. Override the previous rule */
h2 {
  color: blue;
  font-style: italic;
}
  
/* Rule 4: Apply to all the <p> elements */
p {
  text-align: justify;
  color: black;
}"""
```

## I CSS selectors

* ogni elemento del foglio di stile è una **style rule**
* ogni rule è caratterizzata da
    * un **selettore**: definisce l'insieme di elementi **HTML** a cui va applicata la regole di stile
    * una lista di proprietà racchiuse tra parentesi graffe
        * ogni proprietà è una coppia **nome: valore**
        * proprietà relativa al **layout** di quell'elemento/di quegli elementi

## Trovare i riferimenti mediante i css selectors

```python
import requests
from bs4 import BeautifulSoup
url =  'https://en.wikipedia.org/w/index.php' + '?title=List_of_Game_of_Thrones_episodes&oldid=802553687'
r = requests.get(url)
html_contents = r.text
html_soup = BeautifulSoup(html_contents)
for link in html_soup.select('ol.references cite a[href]'):
      print(link.get('href'))
```
