## Approfondimento: utilizzo delle slices

*  definiamo una funzione che abbia lo stesso comportamento della funzione predefinita sum(), su qualsiasi tipo di parametri su cui sia definito l'operatore '+'
* output non può essere impostato a **0**, (non conosciamo il tipo dei dati della sequenza), 
     * assegnamo il primo elemento della sequenza
* il primo elemento va escluso dall'iteratore successivo: utilizziamo una slice
* slide accetta anche indici fuori dal range della sequenza e, in questo caso, restituisce una sequenza vuota.

```python
def mysum(*items):
    if not items: 
        return items
    output = items[0]
    for item in items[1:]:
        output += item  
    return output
```

## Approfondimento: comprehension e generator expressions

* definire una funzione **sum_numbers** che, data una stringa contenenete sottostringhe separate da uno spazio bianco

* esempio: input : **10 abc 20 de44 30 55fg 40** output: **100**

```python
def sum_numbers(numbers):
      return sum(int(number)
             for number in numbers.split()
             if number.isdigit())
print(sum_numbers('10 abc 20 de44 30 55fg 40'))
```

## Ex 3:  Lanci ripetuti di una moneta "fair"

*  10 lanci di una moneta *fair*
* valutare la probabilità di avere x volte testa per $0<x<10$
* spazio campionario
    *tutte le possibili sequenze di "Head" e "Tail"*
* *spazio pesato* in questo caso riportiamo per ogni valore di x il numero di sequenze che riportano esattamente x volte testa (da cui si può dedurre il numero di sequenze che riportano esattamente x volte croce)

## Ex 3: Lanci ripetuti di una moneta "fair"

```python
from collections import defaultdict
from itertools import product
def generate_coin_sample_space(num_flips=10): 
    weighted_sample_space = defaultdict(int)
    for coin_flips in product(['Heads', 'Tails'], repeat=num_flips):
        heads_count = len([outcome for outcome in coin_flips
                                   if outcome == 'Heads']) 
        weighted_sample_space[heads_count] += 1

    return weighted_sample_space

weighted_sample_space = generate_coin_sample_space()
assert weighted_sample_space[10] == 1
assert weighted_sample_space[9] == 10
print (weighted_sample_space)
```

## Ex 3: Lanci ripetuti di una moneta "fair"

```python
def get_matching_event(event_condition, generic_sample_space):
    a= set([key for key in generic_sample_space.keys() 
                if event_condition(key)])
    print (a)
    return a

def compute_event_probability(event_condition, generic_sample_space):
    event = get_matching_event(event_condition, generic_sample_space)   
    event_size = sum(generic_sample_space[outcome] 
                     for outcome in event)
    return event_size / sum(generic_sample_space.values())
```

## Ex 3: Lanci ripetuti di una moneta

```python
def is_in_interval(number, minimum, maximum):
    return minimum <= number <= maximum

prob = compute_event_probability(lambda x: is_in_interval(x, 8, 10),
                                 weighted_sample_space)
print(f"Probability of observing more than 7 heads is {prob}")

prob = compute_event_probability(lambda x: not is_in_interval(x, 3, 7),
                                 weighted_sample_space)
print(f"Probability of observing between 3 and 7 heads is {prob}")
```

## Es 6: Simulare lanci multipli della moneta: NumPy

* vogliamo simulare una lunga sequenza di lanci di moneta
* invece di rappresentare lo spazio campionario come prodotto cartesiano, consideriamo l'output di ogni esperimento come un numero casuale tra 0 e 1
* costruiamo una sequenza di 1000 lanci
* utlizziamo per questo la libreria numpy
    * modulo random della libreria
    * funzione *np.random.randint*
* verifichiamo la legge dei grandi numeri

## Es 6: Analizzare una lunga sequenza di lanci

```python
import numpy as np
np.random.seed(0)
def frequency_heads(coin_flip_sequence):
    total_heads = len([head for head in coin_flip_sequence if head == 1])
    return total_heads / len(coin_flip_sequence)
coin_flips = []
frequencies = []
for _ in range(1000):
    coin_flips.append(np.random.randint(0, 2))
    frequencies.append(frequency_heads(coin_flips))

plt.plot(list(range(1000)), frequencies)
plt.axhline(0.5, color='k')
plt.xlabel('Number of Coin Flips')
plt.ylabel('Head-Frequency')
plt.show()
```

## Assignment 4

* è data una stringa contenente diverse parole seprate da uno spazio
* creare un dizionario in cui le chiavi siano le parole e i valori il numero di vocali in ogni parola
* ad esempio se la stringa è la seguente "this is an easy test" allora il dizionario risultante sarà {'this':1, 'is':1, 'an':1, 'easy':2, 'test':1}

```python
myString = "this is an easy test"
vocali = "aeiou"

def contaVocali(p):
    numero = sum(1 if lettera in vocali else 0 for lettera in p)
    return numero

myDict = {parola:contaVocali(parola) for parola in myString.split()}

myDict
```
