## La tecnologia XPath

* **XPath** indica **XML Path Language** 
    * un query language che può essere utilizzato per descrivere un **path** da un nodo **A** ad un nodo **B** in un documento **XML/HTML**
    * fa riferimento alla rappresentazione interna del documento **HTML**, come **DOM**, **Document Object Model**
* simile ai **CSS selectors**
* vantaggi di **XPath** rispetto ai **CSS selectors**
    * permette di attraversare l'albero in qualsiasi direzione
    * permette trasformazioni dei risultati, prima di restituirli
    * può essere facilmente esteso con ulteriori funzionalità
    * alcune librerie che supportano **CSS selector** convertono i selettori in **XPath**, perchè più veloce e potente

## Selenium Wait Functions

* le operazioni di **wait** consentono di **attendere** che un elemento sia **caricato**, prime di eseguire una qualsiasi azione su tale elemento

* **Selenium** offre il package **Wait** che include tre tipi di **wait**
    * **Explicit Waits**
    * **Implicit Waits**
    * **Fluent Waits**

* possibile utilizzare anche la funzione **sleep** di **Python**
    * aspetta per un intervallo di tempo predeterminato
    * non consigliato perchè imposta un tempo assoluto, non collegato al caricamento degli elementi


## Selenium Implicitly Wait

* impostano un **tempo di attesa globale**
* quando si ricerca un elemento nel **DOM**, e l'elemento non esiste,  il WebDriver interroga la pagina ripetutamente per un certo periodo di tempo
* lo stesso tempo di attesa per **ogni elemento**
    * una sorta di **catch all**, un tempo valido per tutti gli elementi    
* differenza con **time.sleep()**
    * non appena l'elemento è caricato, prosegue con l'esecuzione, invece la **sleep** attende comunque l'intervallo di tempo prestabilito
    * se passa l'intervallo di tempo stabilito, viene sollevata una **NoSuchElementException**
* applicato a tutti i successivi comandi del **WebDriver** e valido per l'intera sessione, **per default vale 0**
* la sintassi è la seguente

<code>
from selenium import webdriver
driver = webdriver.Chrome()
driver.implicitly_wait(10) # seconds
driver.get("http://somedomain/url_that_delays_loading")
myDynamicElement = driver.find_element_by_id("myDynamicElement")
</code>

## Selenium Fluent Wait

* **Fluent Wait** è un'ulteriore  tipo di attesa che aspetta che una condizione specifica venga soddisfatta 
* a differenza delle wait implicite ed esplicite, **Fluent Wait** permette di definire:
    * il tempo massimo di attesa per una condizione.
    * la frequenza di polling (cioè ogni quanto Selenium verifica se la condizione è soddisfatta).
    * la possibilità di ignorare eccezioni specifiche, come NoSuchElementException.
* quindi questo tipo di wait fornisce
    * maggiore flessibilità: si possono impostare intervalli di polling diversi invece di controllare continuamente.
    * migliore gestione degli elementi dinamici: utile per attendere elementi che compaiono a intervalli imprevedibili.
    * evita attese inutili smette di aspettare non appena l'elemento viene trovato.

## Selenium Fluent Wait

````python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

driver = webdriver.Chrome()
wait = WebDriverWait(driver, timeout=10, poll_frequency=1, ignored_exceptions=[])
driver.get("https://www.example.com")
try:
    element = wait.until(EC.visibility_of_element_located((By.TAG_NAME, "h1")))
    print("elemento trovato")
finally:
    # Quit the driver
    time.sleep(10)
    driver.quit()
    
````

## Scrolling page

* Selenium fornisce anche un insieme di funzioni chiamate **Actions**
* queste funzioni permettono un'emulazione completa delle interazioni  degli utenti su una pagina web. 
* queste interazioni  includono 
    * movimenti del mouse
    * pressioni di tasti 
    * scorrimento.
* Per utilizzare queste funzioni, è necessario creare un oggetto **ActionChains**

## Scrolling page

````python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver import Keys
from selenium.webdriver.common.action_chains import ActionChains
import time
driver = webdriver.Chrome()
driver.get('https://scrapingclub.com/exercise/list_infinite_scroll/')
actions = ActionChains(driver)
def scroll_to_bottom():
    # Define the number of times to scroll (adjust as needed)
    scroll_iterations = 10
    for _ in range(scroll_iterations):
        # Scroll down inviando la **END key** al browser
        actions.send_keys(Keys.PAGE_DOWN).perform()
        # Aspetta un poco per garantore che il contenuto sia stato caricato
        time.sleep(15)
````

## Scrolling page

````python
# Call the function to scroll to the bottom
scroll_to_bottom()

captured_dress = driver.find_elements(By.CLASS_NAME, 'p-4')
for dress in captured_dress:
    try:
        description=dress.find_element(By.TAG_NAME, 'h4').text
        print(description)
        price=dress.find_element(By.TAG_NAME, 'h5').text
        print(price)
    except: 
        pass
   
````

## Tecniche di "evasione": User Agent

* uno **User Agent (UA)** è una stringa inviata dal web browser ad un server
* è inserita nell'header **HTTP** ed identifica 
    * tipo del device
    * il tipo del browser
    * la versione
    * il sistema operativo
* utilizzata del web server per identificare il richiedente e inviare il contenuto in modo che sia compatibile con il browser

## Mascheramento dell'IP: Utilizzare SSLProxies.ORG 

````python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import random

proxies = [] # Lista che cnterrà i proxy come dizionari con chiavi ip e porta

# Funzione per la estrazione dei proxy da sslproxies.org
def generate_proxies():
    proxies.clear() # Svuoto la lista dei proxy (nel caso la funzione venga chiamata più volte non si vogliono avere duplicati)
    proxies_req = Request('https://www.sslproxies.org/')
    proxies_req.add_header('User-Agent', ua.random)
    proxies_doc = urlopen(proxies_req).read().decode('utf8')
    soup = BeautifulSoup(proxies_doc, 'html.parser')
    proxies_table = soup.find('table', class_='table table-striped table-bordered')
    # Salvo i proxy nella lista proxies
    for row in proxies_table.tbody.find_all('tr'):
        td = row.find_all('td')
        proxies.append({
        'ip':   td[0].string,
        'port': td[1].string
        })

generate_proxies() 
````

````python
import time
def get_service_name_walletexplorer(address, proxies):
    url = f"https://www.walletexplorer.com/address/{address}"
    original_proxy_count = len(proxies)
    while True:
        if len(proxies) == 0:
            print(f"Numero massimo di proxy bloccati raggiunto ({original_proxy_count}). Terminazione dello script di deanonimizzazione.)")
            raise StopIteration
        # Esegui lo scraping con proxy e User Agent casuali
        proxy = random.choice(proxies)
        print("usedproxie",proxy)
        user_agent = ua.random
        try:
            headers = {'User-Agent': user_agent}
            response = requests.get(url, headers=headers, proxies=proxy)
            print(response)
            soup = BeautifulSoup(response.text, 'html.parser')
            if response.text.startswith("Too"):
                 proxies.remove(proxy)
                 print(f"WalletExplorer ha bloccato l'IP del proxy. Cambio proxy... (numero di proxy rimanenti alla sospensione dell'esecuzione: {len(proxies)})")
                 time.sleep(5)  # Attendi 5 secondi prima di provare un nuovo proxy
                 continue
            wallet_note = soup.find('div', class_='walletnote')
            if wallet_note:
                service_name = wallet_note.find('a').text
                return service_name
            return None        
        except:
            proxies.remove(proxy)
            print(f"Errore durante la richiesta. Cambio proxy... (numero di proxy rimanenti alla sospensione dell'esecuzione: {len(proxies)})")
            time.sleep(5)  # Attendi 5 secondi prima di provare un nuovo proxy
            continue
            
````

## Attenzione agli honeypot

* una tecnica molto diffusa, da parte degli amministratori di applicazioni web, è quella
di inserire volontariamente dei campi all’interno delle form che siano **volutamente nascosti**

* ad esempio

$<$input name="asdfefda2" value="sd32refda" **type="hidden"**$>$

* un utente è in grado di compilare unicamente i campi visualizzati sul proprio schermo
* un bot automatizzato a riempire le form potrebbe riempire anche quelle sezioni dell'honetpot
* se lo scraper invia la form compilata, l’applicazione web immediatamente interrompe la comunicazione con quest’ultimo, precludendo ad esso qualsiasi tipo di accesso.

## HTTP Cookies

* **HTTP** è nato come un protocollo **stateless** 
    * il server non mantiene alcuna informazione sulle richieste precedenti inviate dallo stesso client
    * per introdurre una forma semplice di **stato** è stato introdotto il meccansimo dei **cookies**
* **cookie** (web cookies, HTTP cookies, browser cookies)   
    * un cookie è un dato di piccole dimensioni (pochi kBytes) inviato da un server ad un browser e **memorizzato dal browser** nel computer locale.
    * il browser automaticamente include il cookie in tutte  le successive richieste che fa al medesimo server
        * solo i cookies memorizzati da un browser che sono in relazione con il dominio associato alla URL saranno inviati al server
    * in questo modo il server può identificare in modo univoco un client (un browser)
    * usati per **session identification**, **tracking**, **security**, **shopping cart items**   
* nome **cookie**: un **opaque piece of data held by an intermediary**
    * il dato è di interesse solo per il server e non per il client

## HTTP Cookies

* quando si invia una richiesta HTTP, il server  può includere una richiesta **Set-Cookies** come segue

<code>
HTTP/1.1 200 OK
Content-type: text/html
Set-Cookie: sessionToken=20495; Expires=Wed, 09 Jun 2021 10:10:10 GMT
Set-Cookie: siteTheme=dark
[...]
</code>

* **session token**: identificatore dell'utente che il server usa per riconoscere una visita come appartenente alla stessa sessione
* **siteTheme**: preferenza dell'utente
* **expires**: data oltre la quale il token non è più valido nel server

## HTTP Cookies

* quando il browser riceve l'header **Set-Cookies**
    * memorizza il cookie
    * inserisce il cookie in tutte le richieste successive verso lo stesso browser
    * utilizza l'header **Coookie**
 
<code>
GET /anotherpage.html HTTP/1.1
Host: www.example.com
Cookie: sessionToken=20495; siteTheme=dark
[...]
</code>

## HTTP Cookies and Scraping

````python
import requests
url = 'http://www.webscrapingfordatascience.com/cookielogin/'

# Eseguo una post
# mi loggo sul sito
r = requests.post(url, data={'username': 'Laura', 'password': 'Ricci'})

# Accedo ai cookies ricevuti
receivedcookies=r.cookies
print("questo è il cookie ricevuto")
print(receivedcookies)

# r.cookies is a RequestsCookieJar object which can also
# be accessed like a dictionary.

my_cookies = r.cookies

print("going to the secret page")
r = requests.get(url + 'secret.php', cookies=my_cookies)
print(r.text)
````

## Cookies in Selenium

tramite JavaScript o eventi lato clien
per questi casi, è necessario usare strumenti come Selenium 
* Selenium offre metodi per visualizzare, aggiungere, eliminare e manipolare i cookie.

    * **get_cookies()** recupera tutti i cookie attuali;
    * **add_cookie()** aggiunge un cookie specifico;
    * **delete_all_cookies()** elimina tutti i cookie.
