## Analisi di un dataset reale: Mapping dei valori hash

* lavorare con valori hash può implicare larga occupazione di memoria
* idea: mappare ogni hash diverso in un identificatore unico

```python
import numpy as np
category_map = {'9cea719429e9': 0, np.nan: 1, '188f9a32c360': 2, 
                'ac0b88e46e20': 3}
col_to_mapping = {'Relationship_Status': category_map}

nums = [category_map[hash_code] for hash_code in df_profile.Relationship_Status.values]
df_profile['Relationship_Status'] = nums
print(df_profile.Relationship_Status)
```

```python
print(col_to_mapping)
```

## Analisi di un dataset reale: Mapping dei valori hash

```python
for column in df_profile.columns:
    if column in col_to_mapping:
        continue         
    unique_ids = sorted(set(df_profile[column].values))
    category_map = {id_: i for i, id_ in enumerate(unique_ids)}
    col_to_mapping[column] = category_map    
    nums = [category_map[hash_code] 
            for hash_code in df_profile[column].values]    
    df_profile[column] = nums
head = df_profile.tail()
print(head.to_string(index=False))
```

## Analisi di un Dataset reale: Costruzione del grafo

```python
import networkx as nx
G = nx.Graph()
for id1, id2 in df_friends.values:
    node1 = col_to_mapping['Profile_ID'][id1]
    node2 = col_to_mapping['Profile_ID'][id2]
    G.add_edge(node1, node2)
    
nodes = list(G.nodes)
num_nodes = len(nodes)
print(f"The social graph contains {num_nodes} nodes.")
edges = list(G.edges)
num_edges = len(edges)
print(f"The social graph contains {num_edges} edges.")
```

## Distribuzione dei gradi dei nodi

* Grado di un nodo **v** in un grafo non orientato
  * **$\delta(v)$**: numero di archi incidenti su v.
* un ciclo viene contato due volte.
* nei grafi orientati si distinguono grado entrante (**in-degree**) e grado uscente (**out-degree**).
* l'analisi della distribuzione dei gradi dei nodi fornisce informazioni sulla struttura della rete, ad esempio:
    * se la maggior parte dei vertici ha lo stesso grado (distribuzione uniforme), allora tutti i vertici hanno ruoli simili nella rete.
    * se esistono pochi nodi con grado molto alto, questi sono "hub" e la loro rimozione potrebbe suddividere la rete in più parti.

* strumenti per l'analisi del grado dei nodi:
    * istogrammi: mostrano il numero di vertici con un determinato grado.
    * Sequenze di gradi.
    * Correlazione tra gradi.

## Random Graph:  (Erdos, Renyi, anni'50) 

* un modello di base per l'analisi delle reti complesse
    * consideriamo un insieme di nodi.
    * partiamo da una collezione di n nodi disconnessi.
    * creiamo un collegamento tra ogni coppia di nodi con una probabilità **p**, indipendentemente dagli altri collegamenti.
    * ogni connessione è indipendente dalle altre, seguendo una distribuzione binomiale o di Poisson.

* questo è un buon modello per le reti complesse?
    * sembrava tale fino a circa 20 anni fa.
    * tuttavia, l’evoluzione tecnologica ha:
        * permesso l’analisi sperimentale della struttura delle reti complesse reali, con milioni di nodi.
        * dimostrato che il modello casuale non riflette completamente la struttura di molte reti complesse.

* a allora, perché partire da questo modello?
    * è una buona base per lo studio di modelli più avanzati!


## Random Graph:  (Erdos, Renyi, years'50) 

* **Erdos-Renyi Model**: un grafo non diretto $G_{n,p} (ER(n,p))$, caratterizzato da due parametri
    * **n** numero di nodi
    * **p** probabilità

* procedura di costruzione  
    * per ogni coppia di nodi $(u,v), u \neq v$ tra tutte le possibili coppie $\binom{n}{2}$ = $\frac{n(n-1)}{2}$  “tira” una moneta "truccata"in cui la probabilità di ottenere testa sia **p** 
        * se il risultato è **testa**, aggiungi l'arco tra **u** e **v**
        * se il risultato è  **croce**, non aggiungere l'arco
    * equivalente ad aggiungere la'rco con probabilità  p,

* scelte indipendenti: la presenza/assenza di un arco è indipendente da quella degli altri archi

* un grafo semplice: nessun cappio,  al più un arco tra vertici distinti

## Erdos, Renyi: distribuzione dei gradi dei nodi

* studiamo la **distribuzione del grado dei nodi** di un grafo **non diretto*** da un punto di vista teorico
* calcoliamo la probabilità che il grado di un vertice **u** sia uguale a **k**, per ogni **k** tale che **0 ≤ k ≤ n−1**
    * un nodo ha grado zero se tutti i lanci di moneta risultano **testa**.
    * un nodo ha grado **(n-1)** se tutti i lanci di moneta relativi risultano **croce**.
    * perché un nodo abbia grado **k**, gli **n-1** lanci devono aver dato **k** **testa** e **(n-1-k)** **croce**.
    * poiché la probabilità di ottenere **testa** è **p** e quella di ottenere **croce** è **1-p**, 
    la probabilità di ottenere **k volte testa** e **(n-1-k)  volte croce** è:
      *  $p^k \times (1-p)^{n-1-k}$
    * tuttavia, abbiamo
      *  $\binom{n-1}{k}$
      modi diversi per selezionare **k** vertici distinti tra **n-1**.

* pertanto, la probabilità che il grado di un nodo sia **k** è data da:
    * $\binom{n-1}{k} \times  p^k \times (1-p)^{n-1-k}$

* di conseguenza, la distribuzione dei gradi in un grafo di Erdős-Rényi può essere rappresentata da una **distribuzione binomiale**
* il valore atteso di una distribuzione binomiale è:
    *  $ p \times (n-1)$

* quindi, il grado medio dei nodi in un grafo di Erdős-Rényi con $p = 0,15 e n = 100$ è:
    * $p=0,15, n=100$  = $0,15 \times 100 = 15$

## Erdos, Renyi: Diametro del grafo

* il numero di nodi 
    * cresce esponenzialmente ad ogni passo
    * dopo L passi, può essere approssimato come $K^L$

* quanti passi servono per raggiungere tutti i  $n-1$ nodi, partendo dal nodo centrale?
    * $K^L = n-1$ and $L$ è una approssimazione del diametro
    * $log K^L$ = $log n$  $\Rightarrow$ $L  \times log K$ = $log n$ $\Rightarrow$ $L = log n / log  K$

* il “grado di separazione” di un grafo random  cresce in modo logarotmico rispetto al numero dei nodi
    * diametro basso
    * il modello cattura la proprietà di avere un “basso diametro”, osservata in molte reti sociali 
    * il modello di **ER** è un **buon predittore** del diametro, ma non lo è altrettanto per quanto riguarda il coefficente di clustering
