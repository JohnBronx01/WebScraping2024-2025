## HTML: un linguaggio di markup

* proposto nei primi anni '90  da **Tim Barnes Lee, (CERN)** per scambiare documenti su **Internet**
    * standard attuale **HTML 5: ottobre 2014**
    * non un linguaggio di programmazione, ma adatto alla definizione di contenuto testuale
* effettuare scraping implica reperire e quindi **parsare** una pagina **HTML**

## HTML: concetti di base

* i tag circondano e descrivono il contenuto
    * **tag di apertura:** nome del tag tra parentesi angolari (es. **\<body\>**)
    * **tag di chiusura:** nome con forward slash tra parentesi angolari (es. **\</body\>**)
* **attributi dei tag**:  
    * danno ulteriori informazioni sull'elemento, ad esempio riferite a come l'elemento verrà visualizzato
    * attributi possono essere proprietà da applicare all'argomento e contenute nei **CSS**, **fogli di stile**
    * sono sempre specificati all'interno del **tag di apertura**
    * coppie **nome = valore** 
* **elemento**: qualsiasi informazione compresa tra il tag di apertura e quelli di chiusura

## HTML: tag di base

* tag per la strutturazione del documento   
  * paragraph: *&lt;p&gt;*
  * headers da livello 1 a livello 6: *&lt;h1&gt;,..,&lt;h6&gt;*
  * ordered e unordered lists:   &lt;ol&gt;, &lt;ul&gt;
     * &lt;li&gt;
  * tables: *&lt;table&gt;*
     * *\<th\>*
     * *\<tr\>*
     * *\<td\>*
* hyperlink verso altri documenti **HTML**
     * anchor tag &lt;a&gt;: &lt;a href="https://www.w3schools.com" Web 3 </a>&gt;
* tag per inserire contenuto multimediale/fragmenti di programma, esempio immagini
     * *\<img\>*

## HTML: \<p\>, \<div\>, \<span\>

* sono **contenitori di testo**
    * **block level element**  occupano il 100% dell'elemento che li contiene
    * **inline element** occupano lo spazio necessario per l'elemento che contengoni
* si comportano in modo diverso:
  * **\<p\>** è un **block level element** e lascia spazio prima e dopo la propria chiusura
  * **\<div\>** 
      * è un **block level element**, non lascia spazio prima e dopo la propria chiusura, ma va a capo
      * consente di raggruppare parti del documento (ad esempio più pargrafi ed applicare a tutti una regole di stile)
  * **\<span\>** è  **inline element**  e quindi non va a capo

## Parsare un documento HTML con BeautifulSoup

```python
from bs4 import BeautifulSoup
small_example_html = """ 
         <!DOCTYPE html>
         <html>
            <body>
               <p> Click <a id='info' href='http://www.example.com'>here</a>
                   for more information.
                </p>
            </body>
          </html>"""
small_soup = BeautifulSoup(small_example_html, 'html.parser')
print(type(small_soup))
print(small_soup.prettify())
```

## BeautifulSoup

* la classe **bs4.BeautifulSoup** accetta due parametri per il suo costruttore
    * una stringa che rappresenta il codice **HTML**
    * il tipo di parser che si intende utilizzare per effettuare il parsing della pagina
        * il parametro parser tecnicamente è un argomento di tipo **keyword**, ma viene sollevato un warning se non viene specificato
        * **html.parser** è il **parser standard**, non richiede una installazione "ad hoc" 
* altre opzioni
    * **lxml**, parser etremamente veloce scritto in C, richiede una installazione apposita
    * **html5lib**, parser meno efficiente, ma simula molto bene il comportamento di un web browser, permettendo di parsare anche
    documenti **HTML** estremamente complessi, ma un pò più meno efficiente
    

## Tipi di dato in BeautifulSoup

* **BeautifulSoup** object
    * rappresenta **tutto l'albero** risultante dal parsing di un documento **HTML**
* **Tag** object
    * rappresenta un qualsiasi tag **HTML** del documeto 
    * molte proprietà e metodi associati: nome del tag, attributi,...
* **NavigableString** object
    * rappresentano il testo all'interno di un **tag**, piuttosto che il tag stesso
* **Comment** object
    * HTML comments
    * poco utili per lo scraping
    

## Parsare un documento con BeautifulSoup

* a partire dal documento **HTML**, BeautifulSoup crea un **parse tree** 
* può ricercare contenuti nell'albero parsato
* **ricerca per posizione** 
  * navigare sull'albero utilizzando funzioni per passare ai figli di un nodo o al padre di un nodo
* **ricerca per contenuto** 
  * ricercare contenuti basandosi sul nome dei tag e sui loro attributi

## Ricerca per contenuto

<code>
findAll(tag, attribute, recursive, text, limit, keywords)
find(tag, attribute, recursive, text,  keywords)
</code>

* applicato ad un oggetto di tipo **BeautifulSoup**
* **tag** una stringa che è il nome di un tag, oppure una lista di tag 
* **attributes** un dizionario Python che individua un insieme di attributi del tag selezionato
* **recursive** valore **booleano**
   * se **Falso** si ricercano solo i tag al livello più alto nell'albero del documento, altrimenti
      la ricerca è ricorsiva
   * per default settato a **True**
* **text**  si cerca un match basato sul campo text del tag, piuttosto che attraverso le proprietà del tag
* **limit** indica il numero massimo di match che devono essere restituiti
* **keyword** selezione di tag con un particolare attributo (stesso effetto usando gli attributi)

## Scraping di Tabelle

* trovare il tag **table** all'interno del documento
* effettuare una ricerca ricorsiva a partire da quel tag, sulle righe della tabella
* **find_all** si può applicare anche a oggetti di tipo **tag**
    * prima mi posiziono su un **tag** e poi proseguo con il sottoalbero

## Scraping di Tabelle

```python
from bs4 import BeautifulSoup
import requests 
# first we should find our table object:
soup = BeautifulSoup(table_doc,'html.parser')
table = soup.find('table')
# then we can iterate through each row and extract either header or row values:
header = []
rows = []
print(type(table.find_all('tr')))
for i, row in enumerate(table.find_all('tr')):
    if i == 0:
        header = [el.text.strip() for el in row.find_all('th')]
    else:
        rows.append([el.text.strip() for el in row.find_all('td')])

print(header)
for row in rows:
    print(row)
```

## La funzione enumerate

* aggiunge un contatore a ogni elemento in una lista o in un altro iterabile. 
* ogni elemento viene associato al suo numero (che parte da 0 per impostazione predefinita).
* possibile trasformarlo in una lista di coppie (numero, elemento) utilizzando list().

```python
a = ["WebScraping", "Blockchain", "PeertoPeer"]
print(type(enumerate(a)))
for i, name in enumerate(a):
    print(f"Index {i}: {name}")
print(list(enumerate(a)))
```

## First Scraping example: Scraping Game of Thrones

```python
import requests
from bs4 import BeautifulSoup
url = 'https://en.wikipedia.org/w/index.php?title=List_of:Gam_of_thrones_episodes&oldid=802553687'
r = requests.get(url)
html_contents = r.text
html_soup = BeautifulSoup(html_contents, 'html.parser')
first_h1 = html_soup.find('h1')
print(str(first_h1))
print(first_h1.name) 
print(first_h1.contents) 
```

## First Scraping example: Scraping Game of Thrones

```python
print(first_h1.text) # List of Game of Thrones episodes
print(first_h1.get_text()) # Does the same
print(first_h1.attrs)
```

## First Scraping example: Scraping Game of Thrones, Attributi

```python
print(first_h1.attrs['id']) # firstHeading
print(first_h1['id']) # Does the same
print(first_h1.get('id')) # Does the same
```

## First Scraping example: Scraping Game of Thrones

```python
print('------------ CITATIONS ------------')
# Find the first five cite elements with a citation class
cites = html_soup.find_all('cite', limit=5)
for citation in cites:
    print(citation.get_text())
    # Inside of this cite element, find the first a tag
    link = citation.find('a')
    print(link.text)
    
```
